```
from vecstack import stacking
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score #works
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
#from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.svm import SVC
from collections import Counter #for Smote,

import warnings
warnings.filterwarnings("ignore")
```

```
trainfile = r'/gdrive/My Drive/CIS 508/Assignment 3/RevisedHomesiteTrain.csv'
train_data = pd.read_csv(trainfile)

testfile = r'/gdrive/My Drive/CIS 508/Assignment 3/RevisedHomesiteTest.csv'
test_data = pd.read_csv(testfile)

print(train_data.shape)


print(test_data.shape)


QuoteNum_Data = r'/gdrive/My Drive/CIS 508/Assignment 3/QuoteNumberData.csv'
Quote_Data = pd.read_csv(QuoteNum_Data)

train_data.head()
```

```
#Copy Train data excluding target
trainData_Copy = train_data.iloc[:, :-1].copy()
testData_Copy = test_data.iloc[:, :-1].copy()

X_train = trainData_Copy
X_test = testData_Copy
y_train = train_data['QuoteConversion_Flag']

Quote_Num = Quote_Data[['QuoteNumber']].copy()



```

```
y_train.shape
X_train.shape
```

```
#CONSTRUCT DEFAULT DECISION TREE AND OBTAIN RESPECTIVE ACCURACY ==================
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
clf_predict=clf.predict(X_test)
dt_pred = clf.predict(X_test)
dt_pred = pd.DataFrame(dt_pred,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, dt_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_Decision_Tree_results.csv", index = False)

res_dt = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_Decision_Tree_results.csv")
res_dt.head()






```

```
#Random Forest =============================================================
#Default mode
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_predict=rfc.predict(X_test)
rf_pred = rfc.predict(X_test)
rf_pred = pd.DataFrame(rf_pred,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, rf_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_RandomForest_results.csv", index = False)

res_rf = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_RandomForest_results.csv")
res_rf.head()
```

```
#Gradient Boosting ============================================================
search_grid={'n_estimators':[5,10,20, 30, 50],'learning_rate':[0.01,.1]}
abc =GradientBoostingClassifier()
abc.fit(X_train, y_train)
abc_predict=abc.predict(X_test)

gb_pred = abc.predict(X_test)
gb_pred = pd.DataFrame(gb_pred,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, gb_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_GradientBoosting_results.csv", index = False)

res_gb = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_GradientBoosting_results.csv")
res_gb.head()
```

```
#Construct Multilayer Perceptron using the best parameters
#Default mode
mlp = MLPClassifier()
mlp.fit(X_train, y_train)
mlp_predict=mlp.predict(X_test)


mlp_pred = mlp.predict(X_test)
mlp_pred = pd.DataFrame(mlp_pred,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, mlp_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_MultilayerPerceptron_results.csv", index = False)

res_mlp = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_MultilayerPerceptron_results.csv")
res_mlp.head()
```

```
#Construct Support Vector Machines using the best parameters
#Default mode
svc = svm.SVC()
svc.fit(X_train, y_train)
svc_predict=svc.predict(X_test)

svm_pred = svc.predict(X_test)
svm_pred = pd.DataFrame(svm_pred,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, svm_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_SupportVectorMachines_results.csv", index = False)

res_svm = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_SupportVectorMachines_results.csv")
res_svm.head()
```

```
#SMOTE==============================================================================
print("___________________________________________________________________\nSMOTE\n")
print('Original dataset shape %s' % Counter(y_train))
sm = SMOTE(sampling_strategy='float', ratio=0.5)
X_res, y_res = sm.fit_resample(X_train, y_train)
print('Resampled dataset shape %s' % Counter(y_res))
```

```
#STACKING MODELS =====================================================================
print("___________________________________________________________________________________________\nEnsemble Methods Predictions using DecisionTreeClassifier(), RandomForestClassifier(), GradientBoostingClassifier(), MLPClassifier(), svm.SVC()\n")

models = [ DecisionTreeClassifier(), RandomForestClassifier(), GradientBoostingClassifier(), MLPClassifier()]

S_Train, S_Test = stacking(models,
                           X_res, y_res, X_test,
                           regression=False,

                           mode='oof_pred_bag',

                           needs_proba=False,

                           save_dir=None,

                           metric=accuracy_score,

                           n_folds=4,

                           stratified=True,

                           shuffle=True,

                           random_state=0,

                           verbose=2)
```

```
#STACKING - CONTRUCT A GRADIENT BOOSTING MODEL==============================
model = GradientBoostingClassifier()

model = model.fit(S_Train, y_res)
stack_predict = model.predict(S_Test)

stack_pred = pd.DataFrame(stack_predict,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, stack_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_Stacking_results.csv", index = False)
res_stack = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_Stacking_results.csv")

res_stack.head()
```

```
#Get Prediction Probability for the predicted class as a dataframe
pred_Probability =pd.DataFrame(model.predict_proba(S_Test))

pred_Probability.head()
```

```
#Randomized Search for hyperparameter tuning
model_random = RandomizedSearchCV(model,search_grid,n_iter=15)
model_random.fit(S_Train, y_res)
grid_parm_model=model_random.best_params_
print(grid_parm_model)
```

```
#Construct Gradient Boosting Trees using the best parameters
stack_abc= GradientBoostingClassifier(n_estimators= 10, learning_rate= 0.1)
stack_abc.fit(S_Train,y_res)
stack_abc_predict = stack_abc.predict(S_Test)


#run cross-validation on best parameters, get auc score
stack_abc_cv_score = cross_val_score(stack_abc, S_Train, y_res, cv=10, scoring="roc_auc")
print("=== All AUC Scores ===")
print(stack_abc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Boosting: ",stack_abc_cv_score.mean())


stack_abc_pred = pd.DataFrame(stack_abc_predict,columns=['QuoteConversion_Flag'])

pd.concat([Quote_Num, stack_abc_pred], axis=1).to_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_StackingHyperparamatertuned_results.csv", index = False)
res_stack_tuned = pd.read_csv("/gdrive/My Drive/CIS 508/Assignment 3/Homework3_StackingHyperparamatertuned_results.csv")

res_stack_tuned.head()
```
